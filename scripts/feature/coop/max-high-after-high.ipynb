```python
import psycopg2
import numpy
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
ccursor = COOP.cursor()

highs = numpy.zeros( (2012-1893, 366), 'f')
```

```python
import datetime
ccursor.execute("""SELECT year, extract(doy from day), high from alldata_ia
WHERE station = 'IA2203' and year > 1892 and year < 2012""")
for row in ccursor:
    yindex = int(row[0] - 1893)
    xindex = row[1] -1
    highs[yindex,xindex] = row[2]
```

```python
counts = numpy.zeros((30,), 'f')
sums = numpy.zeros((30,),'f')
mins = numpy.zeros((30,), 'f')
mins[:] = 1000.0
maxs = numpy.zeros((30,), 'f')
maxs[:] = 0.0
for threshold in range(70,100):
    for year in range(1893,2012):
        indices = numpy.where(highs[year-1893,:] >= threshold)
        if len(indices[0]) == 0:
            continue
        lastmax = numpy.max(highs[year-1893,(indices[0][-1]+1):])
        counts[threshold-70] += 1
        sums[threshold-70] += lastmax
        if lastmax < mins[threshold-70]:
            mins[threshold-70] = lastmax
        if lastmax > maxs[threshold-70]:
            maxs[threshold-70] = lastmax
```

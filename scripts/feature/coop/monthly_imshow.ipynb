```python
import numpy.ma as ma
import psycopg2
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

cursor.execute("""
 SELECT year, month, ob, avg(ob) OVER (PARTITION by month ORDER by year ASC ROWS BETWEEN 30 PRECEDING 
 AND CURRENT ROW) from
 (SELECT year, month, sum(precip) as ob from alldata_ia where station = 'IA0200'
 GROUP by year, month) as foo ORDER by year ASC, month ASC
""")

data = ma.zeros(( (2014-1892), 12))

for row in cursor:
    data[ row[0] - 1893, row[1] - 1] = row[2] / row[3] * 100.0
    
data.mask = ma.where(data < 0.0001, True, False)
data[-1,3] = 200.
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
from matplotlib.patches import Rectangle

cmap = cm.get_cmap('seismic_r')
cmap.set_bad('gray')
bounds = [1,25,50,75,95,105,125,150,200,300]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

(fig, ax) = plt.subplots(1,1)

a = ax.imshow(data, cmap=cmap, norm=norm, aspect='auto', interpolation='nearest')
ax.set_xticks(range(0,12))
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'] )
fig.colorbar(a)
ax.set_yticks(range(121,100,-4))
ax.set_yticklabels(range(2014,1993,-4))
ax.set_ylim(122.0,98.5)
ax.set_title("Ames Monthly Precipitation Departures [%]\nfrom 30 year average prior to year")

r = Rectangle((2.5,107.5), 1, 14, lw=3, fc='None', ec='yellow')
ax.add_patch(r)

r = Rectangle((2.5,107.5), 1, 14, lw=1, fc='None', ec='k', zorder=2)
ax.add_patch(r)

r = Rectangle((4.5,117.5), 4, 3, lw=3, fc='None', ec='yellow')
ax.add_patch(r)

r = Rectangle((4.5,117.5), 4, 3, lw=1, fc='None', ec='k', zorder=2)
ax.add_patch(r)
```

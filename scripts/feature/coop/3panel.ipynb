```python
import psycopg2
import numpy
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
ccursor = COOP.cursor()

precip = numpy.zeros( (2013-1893, 123), 'f')
avgt = numpy.zeros( (2013-1893, 123), 'f')
```

```python
import datetime
ccursor.execute("""SELECT day, (high+low)/2.0, precip from alldata_ia
WHERE station = 'IA0000' and month in (5,6,7,8) and year > 1892""")
for row in ccursor:
    yindex = int(row[0].year - 1893)
    xindex = (row[0] - datetime.date(row[0].year, 5, 1)).days
    precip[yindex,xindex] = row[2]
    avgt[yindex,xindex] = row[1]
```

```python
precip_index = numpy.zeros( (2013-1893, 123), 'f')
temp_index = numpy.zeros( (2013-1893, 123), 'f')
for i in range(1,124):
    ytd_precip = numpy.sum( precip[:,0:i], 1)
    avg_precip = numpy.average( ytd_precip )
    std_precip = numpy.std( ytd_precip )
    precip_index[:,i-1] = (ytd_precip - avg_precip) / std_precip
    
    ytd_temp = numpy.average( avgt[:,0:i], 1)
    avg_temp = numpy.average( ytd_temp )
    std_temp = numpy.std( ytd_temp )
    temp_index[:,i-1] = (ytd_temp - avg_temp) / std_temp
```

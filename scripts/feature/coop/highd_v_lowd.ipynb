```python
%matplotlib inline
import psycopg2
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

h = []
l = []
cursor.execute("""
 select a.high - c.high, a.low - c.low from alldata_ia a JOIN climate c 
 on (to_char(c.valid, 'mmdd') = a.sday) WHERE c.station = a.station and 
 a.station = 'IA2203' and a.month = 1 ORDER by a.day ASC
 """)
for row in cursor:
    h.append( row[0] )
    l.append( row[1] )
```

```python
import numpy as np
import numpy.ma as ma

H, xedges, yedges = np.histogram2d(h, l, bins=(30,30),
                                       range=[[-30,30],[-30,30]])
H2 = ma.array(H)
H2.mask = np.where(H < 0.001, True, False)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects

extent = [yedges[0], yedges[-1], xedges[-1], xedges[0]]
(fig, ax) = plt.subplots(1,1)
res = ax.imshow(H2 / 31.0, interpolation='nearest', extent=extent, aspect='auto')
fig.colorbar(res)
ax.grid(True)
ax.set_ylabel("High Temperature Departure $^\circ$F")
ax.set_xlabel("Low Temperature Departure $^\circ$F, X is 26 Jan 2014")
ax.set_ylim(-30,30)
t = ax.text(l[-1],h[-1],'X', color='yellow', fontsize=20)
t.set_path_effects([PathEffects.withStroke(linewidth=2,
                                                 foreground="k")])
ax.set_title("1893-2014 Des Moines January Temperatures Departures\nFrequency (days/month/yr) of Daily High vs Low Departure")
```

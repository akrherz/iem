```python
import psycopg2
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
ccursor = COOP.cursor()
import mx.DateTime
import numpy
```

```python
sigma = []
diff = []
sigma88 = []
diff88 = []
xticks = []
xticklabels = []
for i in range(1,224):
    end = mx.DateTime.DateTime(2012,8,11)
    start = end - mx.DateTime.RelativeDateTime(days=i)
    if start.day == 1:
        xticks.append(i-1)
        xticklabels.append( start.strftime("%b") )
    ccursor.execute("""
    SELECT year, sum(precip) from alldata_ia where station = 'IA0000' and
    sday BETWEEN '%s' and '%s' GROUP by year
    """ % (start.strftime("%m%d"), end.strftime("%m%d")))
    data = numpy.zeros( (ccursor.rowcount,), 'f')
    for row in ccursor:
        data[row[0]-1893] = row[1]
        
    avg = numpy.average(data)
    stddev = numpy.std(data)
    
    sigma.insert(0, (data[2012-1893] - avg) / stddev )
    sigma88.insert(0, (data[1988-1893] - avg) / stddev )
    diff.insert(0, (data[2012-1893] - avg) )
    diff88.insert(0, (data[1988-1893] - avg) )
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(2,1, sharex=True)

ax[0].plot(numpy.arange(223), sigma[::-1], label='2012')
ax[0].plot(numpy.arange(223), sigma88[::-1], label='1988')
ax[0].legend(loc=1, ncol=2)
ax[0].grid(True)
ax[0].set_title("Iowa Accumulated Precipitation Departure [1893-2012]")
ax[0].set_ylabel("$\sigma$ Departure")

ax[1].plot(numpy.arange(223), diff[::-1], label='2012')
ax[1].plot(numpy.arange(223), diff88[::-1], label='1988')
ax[1].legend(loc=1, ncol=2)
ax[1].grid(True)
ax[1].set_xticks(xticks)
ax[1].set_xticklabels(xticklabels)
ax[1].set_xlim(0,224)
ax[1].set_ylabel("Precip Departure [inch]")
ax[1].set_xlabel("Period from 10 August back to date...")

fig.savefig('test.ps')
import iemplot
iemplot.makefeature('test')
```

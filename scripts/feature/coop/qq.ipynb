```python
%matplotlib inline
import psycopg2
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

highs = [0]*13
lows = [0]*13
for i in range(1,13):
    highs[i] = []
    lows[i] = []
    
cursor.execute("select month, high, low from alldata_ia where station = 'IA2203'")
for row in cursor:
    highs[row[0]].append( row[1] )
    lows[row[0]].append( row[2] )
```

```python
import numpy as np
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)
"""
'-'	solid line style
'--'	dashed line style
'-.'	dash-dot line style
':'	dotted line style
'.'	point marker
','	pixel marker
'o'	circle marker
'v'	triangle_down marker
'^'	triangle_up marker
'<'	triangle_left marker
'>'	triangle_right marker
'1'	tri_down marker
'2'	tri_up marker
'3'	tri_left marker
'4'	tri_right marker
's'	square marker
'p'	pentagon marker
'*'	star marker
'h'	hexagon1 marker
'H'	hexagon2 marker
'+'	plus marker
'x'	x marker
'D'	diamond marker
'd'	thin_diamond marker
'|'	vline marker
'_'	hline marker
"""
styles = [':','-','+']

months = ['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
for mo in range(1,13):
    h = np.percentile(highs[mo], range(101))
    l = np.percentile(lows[mo], range(101))
    ax.plot(h, l[::-1], styles[mo%3], label=months[mo], lw=2)
    
ax.legend(loc=3,ncol=2, fontsize=8)
ax.set_xlabel('High Temperature $^\circ$F')
ax.set_ylabel('Low Temperature $^\circ$F')
ax.grid(True)
ax.set_yticks(range(-40,91,10))
ax.set_xticks(range(-20,121,10))
ax.set_title("1880-2012 Des Moines Temperature Percentiles\nEquivalent High and Low Temperature Percentile by Month")

fig.savefig('test.ps')
import iemplot
iemplot.makefeature('test')
```

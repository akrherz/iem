```python
%pylab inline
import psycopg2
pgconn = psycopg2.connect(database='iem', host='iemdb', user='nobody')
cursor = pgconn.cursor()

data = {'DBQ': [0,], 'MSP': [0,], 'ORD': [0,]}
days = {'DBQ': [0,], 'MSP': [0,], 'ORD': [0,]}

cursor.execute("""
  SELECT id, day, (case when snow > 0 then snow else 0 end) 
  from summary s JOIN stations t on (t.iemid = s.iemid)
  WHERE t.id in ('DBQ', 'MSP', 'ORD') and t.network ~* 'ASOS' and s.day > '2013-07-01' ORDER by day ASC
""")
for row in cursor:
    data[row[0]].append( data[row[0]][-1] + row[2] )
    days[row[0]].append( row[1] )
```

```python
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates

(fig, ax) = plt.subplots(1,1)

ax.plot(days['DBQ'][1:], data['DBQ'][1:], lw=2, label='Dubuque 61.7" (38.1")')
ax.plot(days['MSP'][1:], data['MSP'][1:], lw=2, label='Minneapolis 61.5" (49.4")')
ax.plot(days['ORD'][1:], data['ORD'][1:], lw=2, label='Chicago ORD 80.0" (33.8")')
ax.set_xlim(left=datetime.datetime(2013,11,1))
ax.set_title("1 July 2013 - 20 March 2014 Accumulated Snowfall\nnormal accumulation shown in legend parentheses")
ax.set_ylabel("Accumualted Snowfall [inch]")
ax.grid(True)
ax.legend(loc=2, ncol=1, fontsize=10)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%-d %b\n%Y'))
```

```python
%matplotlib inline
import netCDF4
from pyiem.plot import MapPlot
import numpy as np
from scipy.ndimage.filters import minimum_filter, maximum_filter
import datetime
import pytz
```

```python
lats = None
lons = None
def get_pmsl(ts):
    ''' Get the PMSL field '''
    nc = netCDF4.Dataset('%s-000.nc' % (ts.strftime("%Y%m%d%H"),))
    pmsl = nc.variables['Pressure_reduced_to_MSL'][0,:]
    nc.close()
    return pmsl

def get_mins(ts):
    ''' Get the min indices for a given timestamp '''
    global lats, lons
    nc = netCDF4.Dataset('%s-000.nc' % (ts.strftime("%Y%m%d%H"),))
    if lats is None:
        lats = nc.variables['lat'][:]
        lons = nc.variables['lon'][:]
    pmsl = nc.variables['Pressure_reduced_to_MSL'][0,:]
    nc.close()
    mn = maximum_filter(pmsl, size=75)

    return np.nonzero(pmsl == mn)
```

```python
sts = datetime.datetime(2014,1,1)
sts = sts.replace(tzinfo=pytz.timezone("UTC"))
ets = datetime.datetime(2014,1,6)
ets = ets.replace(tzinfo=pytz.timezone("UTC"))
interval = datetime.timedelta(hours=6)
now = sts
total = None
i = 0
while now < ets:
    if now == sts:
        get_mins(now)
    p = get_pmsl(now)
    if total is None:
        total = p
    else:
        total += p
    i += 1
    now += interval
```

```python
print i
m = MapPlot(sector='custom', north=np.max(lats), south=np.min(lats),
            east=np.max(lons), west=np.min(lons), title='1-25 January 2014 NAM (218 Grid) Mean Sea Level Pressure')
m.pcolormesh(lons, lats, total / i / 100.0, np.linspace(1000,1042,22), units='millibars')
m.postprocess(filename='test.png')
m.close()
```

```python
sts = datetime.datetime(2014,1,1)
sts = sts.replace(tzinfo=pytz.timezone("UTC"))
ets = datetime.datetime(2014,1,6)
ets = ets.replace(tzinfo=pytz.timezone("UTC"))
interval = datetime.timedelta(hours=6)
now = sts
indexes = []
i = 0
while now < ets:
    print now
    indexes.append( get_mins(now) )
    m = MapPlot(sector='conus')
    m.pcolormesh(lons, lats, get_pmsl(now) / 100.0, np.linspace(950,1050,25))
    for xidx,yidx in zip(indexes[-1][0], indexes[-1][1]):
        if xidx < 5 or xidx > (sz[0] - 5) or yidx < 5 or yidx > (sz[1] - 5):
            continue
        x,y = m.map(lons[xidx,yidx],lats[xidx,yidx])
        t = m.ax.text(x,y, '%s' % (i,), fontsize=24, color='k', zorder=100)
        t.set_clip_on(True)
    m.postprocess(filename='%s.png' % (now.strftime("%Y%m%d%H"),))
    m.close()
    i += 1
    now += interval
```
